### Rec

```python
#LPIPS no recurrency on single gaussian small 
path1 = "plots/TR_3_LPIPS/train_e.npy"
path2 = "plots/TR_3_LPIPS/test_e.npy"
path3 = "plots/TR_4_LPIPS/train_e.npy"
path4 = "plots/TR_4_LPIPS/test_e.npy"
path5 = "plots/TR_5_LPIPS/train_e.npy"
path6 = "plots/TR_5_LPIPS/test_e.npy"      

data1 = np.load(path1)
data2 = np.load(path2)
data3 = np.load(path3)
data4 = np.load(path4)
data5 = np.load(path5)
data6 = np.load(path6)

fig1, ax1 =  plt.subplots(1,2, figsize = (20,8), )
ax1[0].plot(data1, label = "lr=1e-3")
ax1[1].plot(data2, label = "lr=1e-3")
ax1[0].plot(data3, label = "lr=1e-4")
ax1[1].plot(data4, label = "lr=1e-4")
ax1[0].plot(data5, label = "lr=1e-5")
ax1[1].plot(data6, label = "lr=1e-5")
ax1[0].set_ylim(0, .25)
ax1[1].set_ylim(0, .25)
ax1[0].legend()
ax1[1].legend()
fig1.suptitle("LPIPS", fontsize=15)
ax1[0].set_xlabel("Epoch")
ax1[0].set_ylabel("Loss")
ax1[1].set_xlabel("Epoch")
ax1[1].set_ylabel("Loss")
ax1[0].set_title("Train")
ax1[1].set_title("Test")
plt.savefig("plots/plot1.png")
```
