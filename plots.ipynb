```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
path1 = "plots/arch(Unet)-e(32)-l(MSE)-o(Adam)-lr(1e-05)_test_en.npy"
path2 = "plots/arch(Unet)-e(32)-l(MSE)-o(Adam)-lr(1e-05)_train_en.npy"
path3 = "plots/arch(Unet)-e(32)-l(MSE)-o(Adam)-lr(1e-05)_testn.npy"
path4 = "plots/arch(Unet)-e(32)-l(MSE)-o(Adam)-lr(1e-05)_trainn.npy"
path5 = "plots/arch(Unet)-e(32)-l(MSE)-o(Adam)-lr(1e-05)_test_e.npy"
path6 = "plots/arch(Unet)-e(32)-l(MSE)-o(Adam)-lr(1e-05)_train_e.npy"
path7 = "plots/arch(Unet)-e(32)-l(MSE)-o(Adam)-lr(1e-05)_test.npy"
path8 = "plots/arch(Unet)-e(32)-l(MSE)-o(Adam)-lr(1e-05)_train.npy"


data1 = np.load(path1)
data2 = np.load(path2)
data3 = np.load(path3)
data4 = np.load(path4)
data5 = np.load(path5)
data6 = np.load(path6)
data7 = np.load(path7)
data8 = np.load(path8)


#Convert to pandas dataframe
data3 = pd.DataFrame(data3, columns=['loss'])
data4 = pd.DataFrame(data4, columns=['loss'])
data7 = pd.DataFrame(data7, columns=['loss'])
data8 = pd.DataFrame(data8, columns=['loss'])

#Compute moving average
avg_data3 = data3.rolling(10000, min_periods=1).mean()
avg_data4 = data4.rolling(5000, min_periods=1).mean()
avg_data7 = data7.rolling(10000, min_periods=1).mean()
avg_data8 = data8.rolling(5000, min_periods=1).mean()


#Fig 1: Per Epoch
fig1, ax1 =  plt.subplots(figsize = (20,6))
ax1.plot(data1, color = 'tab:blue', linestyle = (0, (5,1)), label = 'Nrm_Voxel - Test')
ax1.plot(data2, color = 'tab:blue', label = 'Nrm_Voxel - Train')
ax1.plot(data5, color = 'tab:green', linestyle = (0, (5,1)), label = 'Mod_Voxel - Test')
ax1.plot(data6, color = 'tab:green', label = 'Mod_Voxel - Train')
#ax1.axhline(y=0.186, label='0.186 ')
ax1.legend()
ax1.legend()
fig1.suptitle("Loss per Epoch", fontsize=15)
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss (MSE)")

#Fig 2: Per Batch
fig2, ax2 =  plt.subplots(1,2,figsize = (20,6))
ax2[0].plot(avg_data4, label = 'Nrm_Voxel - Train')
ax2[1].plot(avg_data3, label = 'Nrm_Voxel - Test')
ax2[0].plot(avg_data8, label = 'Mod_Voxel - Train')
ax2[1].plot(avg_data7, label = 'Mod_Voxel - Test')
ax2[0].set_ylim(.16, .24)
ax2[1].set_ylim(.16, .24)
ax2[0].legend()
ax2[1].legend()
fig2.suptitle("Loss per batch", fontsize=15)
ax2[0].set_xlabel("Batch (moving average)")
ax2[0].set_ylabel("Loss (MSE)")
ax2[1].set_xlabel("Batch (moving average)")
ax2[1].set_ylabel("Loss (MSE)")
```
